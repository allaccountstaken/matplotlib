## Set up imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Sharing axes

```python
# Read multiple ROC curves data and plot them on the same axes:
fpr_KNN = pd.read_excel('ROC_Curves.xlsx', 'fpr_KNN')
tpr_KNN = pd.read_excel('ROC_Curves.xlsx', 'tpr_KNN')
fpr_MPL = pd.read_excel('ROC_Curves.xlsx', 'fpr_MLP')
tpr_MPL = pd.read_excel('ROC_Curves.xlsx', 'tpr_MLP')
fpr_SGD = pd.read_excel('ROC_Curves.xlsx', 'fpr_SGD')
tpr_SGD = pd.read_excel('ROC_Curves.xlsx', 'tpr_SGD')
fpr_RF = pd.read_excel('ROC_Curves.xlsx', 'fpr_RF')
tpr_RF = pd.read_excel('ROC_Curves.xlsx', 'tpr_RF')
fpr_DT = pd.read_excel('ROC_Curves.xlsx', 'fpr_DT')
tpr_DT = pd.read_excel('ROC_Curves.xlsx', 'tpr_DT')
```

```python
plt.figure(figsize=(15, 9))
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.plot(fpr_KNN, tpr_KNN, label='KNN', color='green', lw=2)
plt.plot(fpr_DT, tpr_DT, label='DecisionTree', color='orange', lw=2)
plt.plot(fpr_RF, tpr_RF, label='RandomForest', color='purple', lw=2)
plt.plot(fpr_MPL, tpr_MPL, label='MLP', color='red', lw=2)
plt.plot(fpr_SGD, tpr_SGD, label='SGD', color='pink', lw=2)
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=14)
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=14)
plt.title('Receiver Operating Characteristics (ROC) Curves', fontsize=20)
plt.legend(loc='best')
plt.xlim(-0.01, 1.01)
plt.ylim(-0.01, 1.01)
plt.show()
```

## Sharing figure

```python
# Basic sharing of axes 
fig = plt.figure()
ax1 = plt.subplot(221) # first plot on 2 x 2 grid
ax2 = plt.subplot(223) # creates one more axes for the second plot in 2 x 2 grid
ax3 = plt.subplot(122) # axes for another plot in 2 x 1 grid, will occupy the full space

ax1.plot([1,2])
ax2.plot([2,1])
ax3.plot([2,4])

plt.show()
```

```python
# More advances axes sharing
wine_quality = pd.read_csv('winequality.csv', delimiter=';')
iris_types = pd.read_csv('iris_dataset.csv', delimiter=',')
iris_mean = iris_types.mean()
iris_std = iris_types.std()

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
plt.subplots_adjust(hspace = 0.5)

axs[0, 0].hist(wine_quality['alcohol'])
axs[0, 0].axvline(x=wine_quality['alcohol'].mean(), linewidth=2, color='r')
axs[0, 0].set_title('Histogram of Alcohol')
axs[0, 0].set_xlabel('Alcohol Bins')
axs[0, 0].set_ylabel('Frequency')

axs[0, 1].boxplot(wine_quality['alcohol'], vert=False,  flierprops=dict(markerfacecolor='r', marker='D'))
axs[0, 1].set_title('Boxplot of Alcohol')

axs[1, 0].scatter(iris_types['petal_length'], iris_types['petal_width'])
axs[1, 0].set(title='Scatter of Iris Types', xlabel='Petal Length', ylabel='Petal Width', )

axs[1, 1].bar(['sepal_l', 'sepal_w', 'petal_l', 'petal_w'], iris_mean, yerr=iris_std, error_kw=dict(ecolor='r'))
axs[1, 1].set(title='Bar plot of Iris Categories', xlabel='Category', ylabel='Category Mean')


plt.show()
```
